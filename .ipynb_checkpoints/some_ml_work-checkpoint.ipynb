```python
import pandas as pd
from sklearn import preprocessing
import sklearn.metrics as metrics

le = preprocessing.LabelEncoder()

data=pd.read_csv("sample_dataset.csv")
data = data[data['selected'] != -1]
print(data.head(10))

uniquejobs = data['job_family'].unique()

print(uniquejobs)

'''
le_job = preprocessing.LabelEncoder()
job_family = data['job_family'].to_list()
le_job.fit(job_family)
data['job_code'] = le_job.transform(job_family)
'''

le_gender = preprocessing.LabelEncoder()
gender = data['gender'].to_list()
le_gender.fit(gender)
data['gender_code'] = le_gender.transform(gender)

#input_columns = ['MCQ_A10','MCQ_A11','MCQ_A13','MCQ_A14','MCQ_A15','MCQ_A18','MCQ_A2','MCQ_A20','MCQ_B7','MCQ_B76','MCQ_B78','MCQ_B79','MCQ_B8','MCQ_B82','MCQ_B85','MCQ_B89','MCQ_B90','MCQ_B91','MCQ_B95','MCQ_B98','job_code','gender_code']
input_columns = ['MCQ_A10','MCQ_A11','MCQ_A13','MCQ_A14','MCQ_A15','MCQ_A18','MCQ_A2','MCQ_A20','MCQ_B7','MCQ_B76','MCQ_B78','MCQ_B79','MCQ_B8','MCQ_B82','MCQ_B85','MCQ_B89','MCQ_B90','MCQ_B91','MCQ_B95','MCQ_B98','gender_code','sentiment']
```

```python
import re
from nltk.sentiment import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()

# Define a function to clean the text
def clean(text):
# Removes all special characters and numericals leaving the alphabets
    text = re.sub('[^A-Za-z]+', ' ', text)
    return text

def is_positive(text) -> bool:
    """True if tweet has positive compound sentiment, False otherwise."""
    if sia.polarity_scores(text)["compound"] > 0:
        return 1
    return 0

data['cleaned_resp'] = data['text_response'].apply(clean)

#sia.polarity_scores("Wow, NLTK is really powerful!")
data['sentiment'] = data['cleaned_resp'].apply(is_positive)
```

```python
from sklearn.model_selection import KFold
from sklearn.naive_bayes import MultinomialNB, BernoulliNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

models = list()
models.append(DecisionTreeClassifier())
models.append(MultinomialNB())
models.append(BernoulliNB())
models.append(RandomForestClassifier())
models.append(KNeighborsClassifier())

for v in uniquejobs:
    data_jb = data[data['job_family']==v]
    
    k_folds = 5
    kfold = KFold(n_splits=k_folds, shuffle=True)
    for fold, (train_ids, test_ids) in enumerate(kfold.split(data_jb)):
        print(f'FOLD {fold} {v}')
        print('--------------------------------')
        #print(train_ids,test_ids)
        #print(data.iloc[train_ids][input_columns])
        x_train = data_jb.iloc[train_ids][input_columns]
        y_train = data_jb.iloc[train_ids]['selected']
        x_test = data_jb.iloc[test_ids][input_columns]
        y_test = data_jb.iloc[test_ids]['selected']

        avgs = pd.DataFrame()
        for i, m in enumerate(models):
            m = m.fit(x_train, y_train)
            train_accuracy = metrics.accuracy_score(y_train,m.predict(x_train))
            preds_test = m.predict(x_test)
            test_accuracy = metrics.accuracy_score(y_test,preds_test)
            avgs[i] = preds_test
            print('Train Accuracy = ', train_accuracy , ', Test Accuracy = ', test_accuracy)
            if i+1 == len(models):
                avgs['means'] = avgs.mean(axis=1)
                avgs.loc[(avgs.means <= 0.5),'means']= 0
                avgs.loc[(avgs.means > 0.5),'means']= 1
                print(avgs.head())
                avgs_accuracy = metrics.accuracy_score(y_test,avgs['means'])
                print('avg_accuracy', avgs_accuracy)
                


```
